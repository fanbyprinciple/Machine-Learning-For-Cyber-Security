# Part 2 working with Hash grams

lets begin by specifying the folders containing our samples, the parameter N and importing a library for hashing and a library to extract N-grams from a string

```python
from os import listdir
from nltk import ngrams
import hashlib

directories = ["Benign_PE_samples", "Malicious_PE_samples"]
N = 2
```

We create a function to read in the bytes of a file and turn these into N-grams

```python
def read_file(file_path):
    binary_file = open(file_path, "rb")
    data = binary_file.read()
    return data

def byte_sequence_to_Ngrams(byte_sequence, N):
    return ngrams(byte_sequence, N)
```

Now we want to hash the Ngrams

```python
def hash_input(inp):
    return int(hashlib.md5(inp).hexdigest(), 16)
```

```python
def make_ngram_hashable(Ngram):
    return bytes(Ngram)
```

The has_file_Ngrams_into_dictionary function takes an Ngram, hashes it and then increments the count in the dictionary of the hash. The reduction module B ensures that there can be no more than B keys in the dictionary

```python
def hash_file_Ngrams_into_dictionary(file_Ngrams,T):
    for Ngram in file_Ngrams:
        hashable_Ngram = make_ngram_hashable(Ngram)
        hashed_and_reduced = hash_input(hashable_Ngram) % B
        T[hashed_and_reduced] = T.get(hashed_and_reduced, 0) + 1
```

We specify the value for B the largest prime number smaller than 2^16 and create an empty dictionary

```python
B = 65521
T = {}
```

We iterate over our files and count theur hashed Ngrams

```python
for dataset_path in directories:
    samples = [ f for f in listdir(dataset_path)]
    for file in samples:
        try:
            file_path = dataset_path + "/" + file
            file_byte_sequence = read_file(file_path)
            file_Ngrams = byte_sequence_to_Ngrams(file_byte_sequence,N)
            hash_file_Ngrams_into_dictionary(file_Ngrams,T)
        except Exception as e:
            print(e)
```

We select the most frequent K1=100 using heapq

```python
K1 = 1000
import heapq

K1_most_common_Ngrams_Using_Hash_Grams = heapq.nlargest(K1, T)
```

Once the top hashed N grams have been selected these make up the feature set. In order to featurize the sample one iterates over its Ngrams , hashes adn

```python
def featurize_sample(file,K1_most_common_Ngrams_Using_Hash_Grams):
    K1 = len(K1_most_common_Ngrams_Using_Hash_Grams)
    fv = K1 * [0]
    file_byte_sequence = read_file(file_path)
    file_Ngrams = byte_sequence_to_Ngrams(file_byte_sequence,N)
    for Ngram in file_Ngrams:
        hashable_Ngram = make_ngram_hashable(Ngram)
        hashed_and_reduced = hash_input(hashable_Ngram) % B
        if hashed_and_reduced in K1_most_common_Ngrams_Using_Hash_Grams:
            index = K1_most_common_Ngrams_Using_Hash_Grams.index(hashed_and_reduced)
            fv[index] += 1
    return fv
```

finally we featurize the data set
